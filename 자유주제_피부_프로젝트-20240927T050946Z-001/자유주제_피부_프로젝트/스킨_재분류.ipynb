#### 턱 jpg 제거

- 턱 크기 x 1000 이상 이미지 삭제

```python
import os
import cv2
import shutil

# 각 폴더 경로 설정
directories = {
    '지성': '/content/train/지성',
    '중성': '/content/train/중성',
    '복합성': '/content/train/복합성',
    '건성': '/content/train/건성'
}

# 새 폴더 경로 설정
new_directories = {
    '지성': '/content/train/지성2',
    '중성': '/content/train/중성2',
    '복합성': '/content/train/복합성2',
    '건성': '/content/train/건성2'
}

# 새 폴더가 없으면 생성
for folder in new_directories.values():
    if not os.path.exists(folder):
        os.makedirs(folder)

# 이미지 삭제 및 새 폴더에 저장하는 함수
def delete_specific_images(original_folder, new_folder):
    for filename in os.listdir(original_folder):
        file_path = os.path.join(original_folder, filename)

        # 이미지 파일만 처리
        if filename.lower().endswith('.jpg') or filename.lower().endswith('.jpeg') or filename.lower().endswith('.png'):
            # '_08'로 끝나는 파일 삭제
            if filename.endswith('_08.jpg'):
                os.remove(file_path)
                print(f"Deleted: {file_path}")
            else:
                # 조건에 맞으면 새로운 폴더에 저장
                new_file_path = os.path.join(new_folder, filename)
                shutil.copy(file_path, new_file_path)
                print(f"Saved in {new_folder}: {filename}")

# 각 폴더에 대해 이미지 삭제 및 저장 처리
for skin_type in directories:
    print(f"Processing {skin_type} folder...")
    delete_specific_images(directories[skin_type], new_directories[skin_type])

print("모든 이미지 삭제 및 저장 처리가 완료되었습니다.")
```

### train - 이마, 볼

```python
import os
import shutil
import zipfile

# 원본 폴더 경로 설정
folders = {
    '건성': '/content/train/건성2',
    '복합성': '/content/train/복합성2',
    '중성': '/content/train/중성2',
    '지성': '/content/train/지성2'
}

# 새로운 최상위 폴더 경로 설정
output_folder = '/content/train2'

# 새로운 최상위 폴더가 없으면 생성
os.makedirs(output_folder, exist_ok=True)

# 이미지 분류 및 저장
for category, folder in folders.items():
    # 각 카테고리 하위 폴더 경로 설정
    forehead_folder = os.path.join(folder, '이마')
    cheek_folder = os.path.join(folder, '볼')

    # 하위 폴더가 없으면 생성
    os.makedirs(forehead_folder, exist_ok=True)
    os.makedirs(cheek_folder, exist_ok=True)

    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)

        # 이미지 파일만 처리
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            if filename.endswith('_01.jpg'):  # 조건: 이름이 _01로 끝나는 경우
                shutil.copy(file_path, os.path.join(forehead_folder, filename))
                print(f"Saved in {forehead_folder}: {filename}")
            else:
                shutil.copy(file_path, os.path.join(cheek_folder, filename))
                print(f"Saved in {cheek_folder}: {filename}")

# 압축 파일 만들기
zip_filename = '/content/train2.zip'
with zipfile.ZipFile(zip_filename, 'w') as zip_file:
    for root, _, files in os.walk(output_folder):
        for file in files:
            file_path = os.path.join(root, file)
            zip_file.write(file_path, os.path.relpath(file_path, output_folder))

print("이미지 분류 및 저장이 완료되었습니다. train2.zip 파일이 생성되었습니다.")
```
